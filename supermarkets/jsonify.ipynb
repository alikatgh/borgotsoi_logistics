```python
import pandas as pd
```

```python
df = pd.read_excel("supermarkets.xlsx")
```

```python
print(df.columns)
```

```python
# Rename the Columns
df.columns = ['id', 'name', 'address']  # Include 'address'
```

```python
# Replace null values with empty strings
df['name'].fillna('', inplace=True)
```

```python
# Generate JSON with appropriate encoding 
supermarkets_json_string = df.to_json(orient='records', force_ascii=False)

print(supermarkets_json_string) 
```

```python
# Load the JSON data into a DataFrame
df_json = pd.read_json(supermarkets_json_string)
```

```python
# Remove entries with missing addresses
df_json = df_json.dropna(subset=['address'])

# Convert DataFrame back to JSON
df_json = df_json.to_json(orient='records', force_ascii=False)
```

```python
df_json
```

```python
import json
```

```python
# Save as a JSON file with correct encoding
with open('supermarkets.json', 'w', encoding='utf-8') as out_file:
    json.dump(df_json, out_file, indent=4, ensure_ascii=False)
```
